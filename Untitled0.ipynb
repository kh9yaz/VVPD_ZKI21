# Numpy


```python
import numpy as np

# Создание вектора x с 150 элементами, равномерно распределенными в интервале [-10, 10]
x = np.linspace(-10, 10, 150)

# Вычисление значения функции y = sin(x) + N(0, 0.1)
noise = np.random.normal(0, 0.1, size=len(x))  # Генерация шума N(0, 0.1)
y = np.sin(x) + noise

# Поиск наименьшего и наибольшего значений y и соответствующих точек x
min_y = np.min(y)
min_x = x[np.argmin(y)]
max_y = np.max(y)
max_x = x[np.argmax(y)]

# Вывод результатов
print(f"Наименьшее значение y: {min_y}, достигается в точке x: {min_x}")
print(f"Наибольшее значение y: {max_y}, достигается в точке x: {max_x}")
```

# Matplotlib

```python
import matplotlib.pyplot as plt

# Построение точечного графика (scatter plot)
plt.scatter(x, y, color='blue', label='Точки (x, y)', alpha=0.7)
plt.title("Точечный график")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Сортировка данных для построения графика функции
sorted_indices = np.argsort(x)
x_sorted = x[sorted_indices]
y_sorted = y[sorted_indices]

# Построение графика функции (plot)
plt.plot(x_sorted, y_sorted, color='red', label='График y(x)', linewidth=2)
plt.title("График функции y(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
